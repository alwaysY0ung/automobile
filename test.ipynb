```python
import pandas as pd
import cantools
from pathlib import Path
from typing import Union, Optional

class MinMaxScalerFromDBC:
    """
    DBC 파일의 Min/Max 정보를 기반으로 데이터를 스케일링하는 클래스입니다.
    또한, DBC에 직접 기재된 Min/Max와 scale/offset으로 계산된 Min/Max 값의
    불일치를 검사하는 기능도 포함합니다.
    """
    def __init__(self, dbc_file_path: Union[str, Path] = 'datasets/hyundai_2015_ccan.dbc'):
        self.dbc_file_path = Path(dbc_file_path)
        self.db = None
        self.signal_ranges = {}
        self._load_dbc_and_parse_ranges()

    def _load_dbc_and_parse_ranges(self):
        """
        DBC 파일을 로드하고 각 신호의 Min/Max 값을 파싱합니다.
        스케일링에 사용할 범위를 self.signal_ranges 딕셔너리에 저장합니다.
        """
        try:
            self.db = cantools.database.load_file(str(self.dbc_file_path))
            print(f"DBC 파일 '{self.dbc_file_path}' 로드 완료.")
        except Exception as e:
            print(f"오류: DBC 파일을 로드할 수 없습니다. 경로를 확인하세요: {self.dbc_file_path}. 오류: {e}")
            self.db = None
            return

        for message in self.db.messages:
            message_id = message.frame_id
            for signal in message.signals:
                column_name = f"{message_id:03X}_{signal.name}"

                # DBC에 명시된 Min/Max 값이 있으면 이를 우선적으로 사용
                if signal.minimum is not None and signal.maximum is not None:
                    self.signal_ranges[column_name] = {
                        'min': signal.minimum,
                        'max': signal.maximum
                    }
                else:
                    # 명시된 Min/Max가 없는 경우, scale과 offset으로 계산
                    if signal.scale is not None and signal.offset is not None and signal.length is not None:
                        if signal.is_signed:
                            raw_min = -(2**(signal.length - 1))
                            raw_max = (2**(signal.length - 1)) - 1
                        else:
                            raw_min = 0
                            raw_max = (2**signal.length) - 1

                        calculated_min = raw_min * signal.scale + signal.offset
                        calculated_max = raw_max * signal.scale + signal.offset

                        self.signal_ranges[column_name] = {
                            'min': calculated_min,
                            'max': calculated_max
                        }
                    else:
                        print(f"경고: 신호 '{column_name}'는 Min/Max 정보가 부족하여 스케일링 범위 설정 불가.")
                        self.signal_ranges[column_name] = {'min': 0.0, 'max': 1.0}


    def scale(self, df: pd.DataFrame) -> pd.DataFrame:
        """
        데이터프레임을 DBC 파일의 Min/Max 정보를 사용하여 [0, 1] 범위로 스케일링합니다.
        """
        df_scaled = df.copy()

        label_column = None
        if 'label' in df_scaled.columns:
            label_column = df_scaled['label']
            df_scaled = df_scaled.drop(columns=['label'])
            print("'label' 컬럼은 스케일링에서 제외됩니다.")

        columns_to_scale = [col for col in df_scaled.columns if col in self.signal_ranges]
        missing_columns = [col for col in df_scaled.columns if col not in self.signal_ranges]

        if missing_columns:
            print(f"경고: 다음 컬럼들은 DBC 파일에서 Min/Max 정보를 찾을 수 없어 스케일링에서 제외됩니다: {missing_columns}")

        for col in columns_to_scale:
            min_val = self.signal_ranges[col]['min']
            max_val = self.signal_ranges[col]['max']

            if max_val == min_val:
                df_scaled[col] = 0.0
                print(f"경고: 컬럼 '{col}'의 Min/Max 값이 동일하여 0으로 스케일링되었습니다. (Min: {min_val}, Max: {max_val})")
            else:
                df_scaled[col] = (df_scaled[col] - min_val) / (max_val - min_val)

        if label_column is not None:
            df_scaled['label'] = label_column
            if pd.api.types.is_float_dtype(df_scaled['label']) and (df_scaled['label'] == df_scaled['label'].astype(int)).all():
                 df_scaled['label'] = df_scaled['label'].astype(int)
            elif pd.api.types.is_integer_dtype(label_column):
                 df_scaled['label'] = df_scaled['label'].astype(int)
            print("'label' 컬럼이 다시 데이터프레임에 추가되었습니다.")

        return df_scaled

    def check_discrepancies(self):
        """
        DBC 파일에 직접 기재된 Min/Max 값과 scale 및 offset으로 계산된 Min/Max 값을 비교하여
        차이가 나는 경우를 찾아 출력합니다.
        지정된 신호 목록에 있는 경우에만 출력합니다.
        """
        if not self.db:
            print("DBC 파일이 로드되지 않았습니다. 먼저 DBC 파일을 로드하세요.")
            return

        # 필터링할 신호 목록
        allowed_signals = {
            '044_CR_Datc_OutTempC', '044_CF_Datc_IncarTemp', '080_PV_AV_CAN', '080_N', '080_TQI_ACOR',
            '080_TQFR', '080_TQI', '081_BRAKE_ACT', '081_CF_Ems_EngOperStat', '081_CR_Ems_IndAirTemp',
            '111_TQI_TCU_INC', '111_SWI_GS', '111_TQI_TCU', '111_SWI_CC', '112_VS_TCU', '112_N_INC_TCU',
            '112_VS_TCU_DECIMAL', '113_SLOPE_TCU', '113_CF_Tcu_TarGr', '113_CF_Tcu_ShfPatt', '113_CF_Tcu_ActEcoRdy',
            '162_Clutch_Driving_Tq', '162_Cluster_Engine_RPM', '162_Cluster_Engine_RPM_Flag', '18F_R_PAcnC', '18F_TQI_B',
            '18F_R_NEngIdlTgC', '200_FCO', '220_LAT_ACCEL', '220_LONG_ACCEL', '220_CYL_PRES', '220_YAW_RATE',
            '251_CR_Mdps_StrTq', '251_CR_Mdps_OutTq', '260_TQI_MIN', '260_TQI', '260_TQI_TARGET', '260_TQI_MAX',
            '260_CF_Ems_AclAct', '2B0_SAS_Angle', '2B0_SAS_Speed', '316_PUC_STAT', '316_TQI_ACOR', '316_N', '316_TQI',
            '316_TQFR', '316_VS', '329_TEMP_ENG', '329_MAF_FAC_ALTI_MMV', '329_CLU_ACK', '329_BRAKE_ACT', '329_TPS',
            '329_PV_AV_CAN', '381_CR_Mdps_StrAng', '383_CR_Fatc_OutTemp', '383_CR_Fatc_OutTempSns', '386_WHL_SPD_FL',
            '386_WHL_SPD_FR', '386_WHL_SPD_RL', '386_WHL_SPD_RR', '387_WHL_PUL_FL', '387_WHL_PUL_FR', '387_WHL_PUL_RL',
            '387_WHL_PUL_RR', '47F_ROL_CNT_ESP', '4F1_CF_Clu_VanzDecimal', '4F1_CF_Clu_Vanz', '4F1_CF_Clu_DetentOut',
            '50C_CF_Clu_AvgFCI', '50C_CF_Clu_DTE', '52A_CF_Clu_VehicleSpeed', '541_CF_Gway_TurnSigLh',
            '541_CF_Gway_HeadLampLow', '541_CF_Gway_TurnSigRh', '545_AMP_CAN', '545_BAT_Alt_FR_Duty',
            '545_VB', '545_TEMP_FUEL', '547_ECGPOvrd', '547_IntAirTemp', '549_BAT_SNSR_I', '549_BAT_SOC',
            '549_BAT_SNSR_V', '549_BAT_SNSR_Temp', '553_CF_Gway_AutoLightValue', '553_CF_Gway_AvTail',
            '553_CF_Gway_ExtTailAct', '553_CF_Gway_IntTailAct', '555_CR_Fpcm_LPActPre', '556_PID_04h',
            '556_PID_05h', '556_PID_0Ch', '556_PID_0Dh', '556_PID_11h', '557_PID_0Bh', '557_PID_23h',
            '58B_CF_Lca_Stat', '58B_CF_Lca_IndLeft', '58B_CF_Lca_IndRight', '58B_CF_Lca_IndBriLeft',
            '58B_CF_Lca_IndBriRight', '593_PRESSURE_FL', '593_PRESSURE_FR', '593_PRESSURE_RL',
            '593_PRESSURE_RR', '5A0_CF_Acu_Dtc', '5B0_CF_Clu_Odometer'
        }

        print("\n--- 지정된 Min/Max 값 불일치 검사 시작 ---")
        discrepancies_found = False
        tolerance = 1e-9  # 부동 소수점 비교를 위한 허용 오차

        for message in self.db.messages:
            for signal in message.signals:
                column_name = f"{message.frame_id:03X}_{signal.name}"

                # 지정된 신호 목록에 포함된 경우에만 처리
                if column_name in allowed_signals:
                    # 1. DBC 파일에 직접 기재된 Min/Max 값
                    dbc_min = signal.minimum
                    dbc_max = signal.maximum

                    # 2. scale과 offset으로 계산된 Min/Max 값
                    calculated_min, calculated_max = None, None
                    if signal.scale is not None and signal.offset is not None and signal.length is not None:
                        if signal.is_signed:
                            raw_min_val = -(2**(signal.length - 1))
                            raw_max_val = (2**(signal.length - 1)) - 1
                        else:
                            raw_min_val = 0
                            raw_max_val = (2**signal.length) - 1

                        calculated_min = raw_min_val * signal.scale + signal.offset
                        calculated_max = raw_max_val * signal.scale + signal.offset

                    # 3. 값 비교
                    min_differs = False
                    max_differs = False

                    if dbc_min is not None and calculated_min is not None and abs(dbc_min - calculated_min) > tolerance:
                        min_differs = True
                    if dbc_max is not None and calculated_max is not None and abs(dbc_max - calculated_max) > tolerance:
                        max_differs = True

                    # DBC에만 값이 있거나, 계산된 값만 있는 경우도 불일치로 간주
                    if (dbc_min is None and calculated_min is not None) or \
                       (dbc_min is not None and calculated_min is None):
                        min_differs = True
                    if (dbc_max is None and calculated_max is not None) or \
                       (dbc_max is not None and calculated_max is None):
                        max_differs = True

                    if min_differs or max_differs:
                        discrepancies_found = True
                        print(f"신호: {column_name}")
                        print(f"  DBC 기재 Min: {dbc_min}, Max: {dbc_max}")
                        print(f"  계산된 Min: {calculated_min}, Max: {calculated_max}")
                        print("-" * 30)

        if not discrepancies_found:
            print("지정된 신호 목록에서 Min/Max 값의 불일치가 발견되지 않았습니다.")
        print("--- 지정된 Min/Max 값 불일치 검사 완료 ---")


if __name__ == "__main__":
    dbc_file_path = 'datasets/hyundai_2015_ccan.dbc'
    scaler = MinMaxScalerFromDBC(dbc_file_path=dbc_file_path)
    scaler.check_discrepancies()
```
